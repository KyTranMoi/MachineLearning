```python
import requests
from urllib3.util import Retry
from requests.adapters import HTTPAdapter


def get_ip():
    response = requests.get('https://api64.ipify.org?format=json')
    data = response.json()
    current_ip = data['ip']
    return current_ip


# Danh sách các địa chỉ IP bạn muốn sử dụng
ip_addresses = [get_ip() for _ in range(10)]

# Tạo một đối tượng Retry để quy định số lần thử lại và các khoảng thời gian giữa các lần thử lại
retry_strategy = Retry(
    total=5,
    backoff_factor=1,
    status_forcelist=[429, 500, 502, 503, 504],
)

# Tạo một đối tượng HTTPAdapter để sử dụng Retry Strategy
adapter = HTTPAdapter(max_retries=retry_strategy)

# Tạo một Session và thêm adapter đã tạo
session = requests.Session()
session.mount('http://', adapter)
session.mount('https://', adapter)

# Lặp qua danh sách các địa chỉ IP và thực hiện các yêu cầu sử dụng từng địa chỉ IP
for ip in ip_addresses:
    url = f'http://{ip}/github.com'
    
    try:
        response = session.get(url)
        # Xử lý dữ liệu trả về ở đây
        print(f"Response from {ip}: {response.text}")
    except requests.RequestException as e:
        print(f"Error with {ip}: {e}")

# Đừng quên đóng Session khi đã sử dụng xong
session.close()
```

```python
# Given data
data = [6, 7, 15, 36, 39, 40, 41, 42, 43, 47, 49]

# Function to calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# Function to calculate median
def calculate_median(data):
    n = len(data)
    middle = n // 2
    if n % 2 == 0:
        return (data[middle - 1] + data[middle]) / 2
    else:
        return data[middle]

# Function to calculate midrange
def calculate_midrange(data):
    return (min(data) + max(data)) / 2

# Function to calculate quartiles and IQR
def calculate_quartiles(data):
    n = len(data)
    q1 = data[(n + 1) // 4 - 1]
    q2 = calculate_median(data)
    q3 = data[(3 * (n + 1)) // 4 - 1]
    iqr = q3 - q1
    return q1, q2, q3, iqr

# Calculate and print the results
mean_result = calculate_mean(data)
median_result = calculate_median(data)
midrange_result = calculate_midrange(data)
q1_result, q2_result, q3_result, iqr_result = calculate_quartiles(data)

print("Mean =", mean_result)
print("Median =", median_result)
print("Midrange =", midrange_result)
print("Q1 =", q1_result)
print("Q2 =", q2_result)
print("Q3 =", q3_result)
print("IQR =", iqr_result)
```
